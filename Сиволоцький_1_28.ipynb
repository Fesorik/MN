```python
import numpy as np
```

1. Створіть дві матриці розміром 2x2, заповніть їх випадковими дійсними числами в 
діапазоні від 0 до 1 та знайдіть їх добуток.


```python
# Створення матриць
matrix_2x2_a = np.random.rand(2, 2)
matrix_2x2_b = np.random.rand(2, 2)
```

```python
# Обчислення добутку
product_2x2 = np.dot(matrix_2x2_a, matrix_2x2_b)
print("Матриця A:\n", matrix_2x2_a)
print("Матриця B:\n", matrix_2x2_b)
print("Добуток матриць A і B:\n", product_2x2)
```
